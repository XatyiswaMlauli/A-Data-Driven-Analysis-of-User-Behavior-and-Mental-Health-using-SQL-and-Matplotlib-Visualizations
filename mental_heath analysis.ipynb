```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Python EDA environment ready")
```

```python
df = pd.read_csv("mental_health - mental_health.csv")

df.head()
df.info()
df.describe()
```

```python
#Remove duplicates
df = df.drop_duplicates(subset="User_ID")

# Validate age range
df = df[(df["Age"] >= 18) & (df["Age"] <= 60)]
```

```python
#Show whether higher screen exposure aligns with increased anxiety symptoms. Screen-time vs Anxiety
plt.figure()
sns.scatterplot(
    data=df,
    x="Daily_Screen_Time_Hours",
    y="GAD_7_Score"
)
plt.title("Daily Screen Time vs Anxiety (GAD-7 Score)")
plt.xlabel("Daily Screen Time (Hours)")
plt.ylabel("GAD-7 Anxiety Score")
plt.show()
```

```python
#Highlight differences between passive scrolling and active engagement. Depression by Activity Type
plt.figure()
sns.boxplot(
    data=df,
    x="Activity_Type",
    y="PHQ_9_Score"
)
plt.title("Depression Severity by Social Media Activity Type")
plt.xlabel("Activity Type")
plt.ylabel("PHQ-9 Depression Score")
plt.show()
```

```python
#Anxiety by Late-Night Usage 
plt.figure()
sns.barplot(
    data=df,
    x="Late_Night_Usage",
    y="GAD_7_Score"
)
plt.title("Average Anxiety Score by Late-Night Social Media Usage")
plt.xlabel("Late-Night Usage")
plt.ylabel("Average GAD-7 Score")
plt.show()
```
